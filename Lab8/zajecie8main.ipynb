```python
import numpy as np

# Definicja parametrów MDP
states = ["H", "M", "S"]
actions = ["OB", "AB", "HO"]
gamma = 0.95

# Prawdopodobieństwa przejść
transitions = {
    "M": {
        "OB": [(0.60, "M"), (0.30, "S"), (0.10, "H")],
        "AB": [(0.55, "H"), (0.40, "M"), (0.05, "S")],
        "HO": [(0.70, "H"), (0.25, "M"), (0.05, "S")],
    },
    "S": {
        "OB": [(0.75, "S"), (0.20, "M"), (0.05, "H")],
        "AB": [(0.55, "M"), (0.25, "H"), (0.20, "S")],
        "HO": [(0.55, "M"), (0.35, "H"), (0.10, "S")],
    },
    "H": {
        "OB": [(0.90, "H"), (0.10, "M")],
        "AB": [(0.92, "H"), (0.08, "M")],
        "HO": [(0.93, "H"), (0.07, "M")],
    },
}


# Funkcja nagrody
def get_reward(next_state, action):
    reward = 0
    if next_state == "H":
        reward += 9
    elif next_state == "S":
        reward -= 7

    if action == "OB":
        reward += 0
    elif action == "AB":
        reward -= 2
    elif action == "HO":
        reward -= 4

    return reward


# Algorytm Value Iteration
def value_iteration(theta=1e-6):
    V = {s: 0.0 for s in states}
    policy = {s: "OB" for s in states}

    while True:
        delta = 0
        for s in states:
            v = V[s]
            action_values = {}

            for a in actions:
                val = 0
                for prob, next_s in transitions[s][a]:
                    r = get_reward(next_s, a)
                    val += prob * (r + gamma * V[next_s])
                action_values[a] = val

            # Poprawka rozwiązująca błąd "No overloads for 'max'"
            best_action = max(action_values, key=lambda a: action_values[a])
            V[s] = action_values[best_action]
            policy[s] = best_action

            delta = max(delta, abs(v - V[s]))

        if delta < theta:
            break

    return V, policy


if __name__ == "__main__":
    V, optimal_policy = value_iteration()

    print("--- Wyniki Value Iteration ---")
    print("Optymalna polityka:")
    for s in states:
        print(f"Stan {s}: {optimal_policy[s]}")

    print("\nWartości stanów (V):")
    for s in states:
        print(f"V({s}) = {V[s]:.2f}")
```
